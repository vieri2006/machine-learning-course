# Settings
Set the settings for the output images. This is the only part where you need to edit. After that, just run the rest of the code.

```
img_dir = "/content/drive/My Drive/Colab Notebooks/train_prep_image/"
label = "False/" # "True/" or "False/"
height = 300 # or any positive integer
width = 225 # or any positive integer as long as it is smaller than the height
color_mode = "RGB/" # "RGB/" or "BW/"
edge_detection = "normal/" # "canny_edge/" or "normal/"
threshold = 0 # or any positive real number (doesn't matter if you choose "normal/")
```

# Libraries & Functions

```
import numpy as np
import argparse
import glob
import cv2
import os
from os import path

def auto_canny(image, sigma):
    # default sigma = 0.33
    # compute the median of the single channel pixel intensities
    v = np.median(image)
 
    # apply automatic Canny edge detection using the computed median
    lower = int(max(0, (1.0 - sigma) * v))
    upper = int(min(255, (1.0 + sigma) * v))
    edged = cv2.Canny(image, lower, upper)
 
    # return the edged image
    return edged
```

# Image Preprocessing
**DO NOT EDIT UNLESS YOU KNOW WHAT YOU ARE DOING**

```
dim = str(height) + "x" + str(width) + "/"
i = 0

for imagePath in os.listdir(img_dir + label + "800x600/RGB/normal/"):  # reading all images from the original dataset one by one
  # get the image
  image = cv2.imread(img_dir + label + "800x600/RGB/normal/" + imagePath)

  # Pixel Reduction
  if dim == "800x600/":
    pass
  else:
    (H, W) = image.shape[:2]
    if H >= W:
      new_h = height
      new_w = width  # round(new_h*W/H)
    else:
      new_w = height
      new_h = width  # round(new_w*H/W)
    image = cv2.resize(image, (new_w, new_h))

  # Change Color Mode
  if color_mode == "RGB/":
    pass
  else:
    image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

  # Choose to apply edge detection or not
  if edge_detection == "normal/":
    thresh = ""
  else:
    blurred = cv2.GaussianBlur(image, (3, 3), 0)
    image = auto_canny(blurred, threshold)
    thresh = str(threshold) + "/"

  # Save the edited image to the correct directory
  if path.exists(img_dir + label + dim + color_mode + edge_detection + thresh):
    cv2.imwrite(img_dir + label + dim + color_mode + edge_detection + thresh + str(i) + ".jpg", image)
  else:
    os.makedirs(img_dir + label + dim + color_mode + edge_detection + thresh)
    cv2.imwrite(img_dir + label + dim + color_mode + edge_detection + thresh + str(i) + ".jpg", image)

  i += 1
```
